# Import modules

```python
import csv
import numpy as np
import pandas as pd
from tqdm.notebook import tqdm
from polyglot.mapping import Embedding, CaseExpander, DigitExpander
```

# Read Dataset and Word Embeddings

```python
raw_dataset = pd.read_csv("../datasets/raw/Indonesian_Manually_Tagged_Corpus.tsv", sep="\t", header=None, names=["token", "tag"], quoting=csv.QUOTE_NONE, skip_blank_lines=False)
```

# Number of sentence

```python
tokens = [token for token in raw_dataset.token.values]

def number_of_sentence(tokens, return_sentences=False):
    sentence = []
    sentences = []

    for token in tokens:
        if token is not np.nan:
            sentence.append(token)
        else:
            sentences.append(sentence)
            sentence = []
    
    if return_sentences:
        return len(sentences), sentences 
    
    return len(sentences)

n_sentence, sentences = number_of_sentence(tokens, True)
print(f"Number of sentence: {n_sentence}")
print(f"Sampel sentence: {sentences[:3]}")
```

# Number of Token

```python
words = [token for token in tokens if token is not np.nan]
print(f"Number of word: {len(words)}")
```

# Number of Vocab

```python
def number_of_vocab(tokens, lowercase=False, return_vocab=False):
    if lowercase:
        vocabs = list(set([token.lower() for token in tokens if token is not np.nan]))
    else:
        vocabs = list(set([token for token in tokens if token is not np.nan]))
    
    if return_vocab:
        return len(vocabs), vocabs
    
    return len(vocabs)

n_vocab, vocabs = number_of_vocab(tokens=tokens, lowercase=False, return_vocab=True)
print(f"Number of vocab without lowercase: {n_vocab}")
print(f"Vocabs: {vocabs[:10]}\n")

n_lower_vocab, lower_vocabs = number_of_vocab(tokens=tokens, lowercase=True, return_vocab=True)
print(f"Number of vocab with lowercase: {n_lower_vocab}")
print(f"Vocabs: {lower_vocabs[:10]}")
```

# Shortest and Longest Sentence

```python
def shortest_item(items):
    items = [item for item in items if item is not np.nan]
    len_items = [len(item) for item in items if item is not np.nan]
    idx_shortest_item = len_items.index(min(len_items))
    shortest_item = items[idx_shortest_item]
    len_shortest_item = len(shortest_item)

    return idx_shortest_item, shortest_item, len_shortest_item

def longest_item(items):
    items = [item for item in items if item is not np.nan]
    len_items = [len(item) for item in items if item is not np.nan]
    idx_longest_item = len_items.index(max(len_items))
    longest_item = items[idx_longest_item]
    len_longest_item = len(longest_item)

    return idx_longest_item, longest_item, len_longest_item

idx_shortest_sentence, shortest_sentence, len_shortest_sentence = shortest_item(sentences)
idx_longest_sentence, longest_sentence, len_longest_sentence = longest_item(sentences)

idx_shortest_token, shortest_token, len_shortest_token = shortest_item(tokens)
idx_longest_token, longest_token, len_longest_token = longest_item(tokens)

print(f"Index shortest sentence: {idx_shortest_sentence}")
print(f"Shortest sentence: {shortest_sentence}")
print(f"Length shortest sentence: {len_shortest_sentence}\n")

print(f"Index longest sentence: {idx_longest_sentence}")
print(f"Longest sentence: {longest_sentence}")
print(f"Length longest sentence: {len_longest_sentence}\n")

print(f"Index shortest token: {idx_shortest_token}")
print(f"Shortest token: {shortest_token}")
print(f"Length shortest token: {len_shortest_token}\n")

print(f"Index longest token: {idx_longest_token}")
print(f"Longest token: {longest_token}")
print(f"Length longest token: {len_longest_token}")
```

# Out of Vocabulary

## Load Word Embeddings

```python
idn_word_embeddings = Embedding.load("../word_embeddings/polyglot/idn_embeddings.tar.bz2")
idn_word_embeddings.apply_expansion(DigitExpander)
idn_word_embeddings.apply_expansion(CaseExpander)
```

```python
def oov_token(vocabs, word_embeddings, return_oov_token=False):
    oov = [vocab for vocab in vocabs if vocab not in word_embeddings]
    
    if return_oov_token:
        n_oov = len(oov)
        
        return n_oov, oov
    
    return len(oov)

import string

chars = list(string.ascii_letters + string.digits + string.punctuation)
n_oov, oov = oov_token(vocabs + chars, idn_word_embeddings, True)
```

```python
print(f"Number of not OOV: {n_vocab - n_oov} ({round(100 - (n_oov / n_vocab * 100), 2)}%)")
print(f"Number of OOV: {n_oov} ({round(n_oov / n_vocab * 100, 2)}%)")
```

```python
def sentence_with_oov(sentences, oov):
    oov_sentences = []
    
    for sentence in tqdm(sentences):
        n_oov_token_in_sent = 0
        
        for token in sentence:
            if token in oov:
                n_oov_token_in_sent += 1
        
        if n_oov_token_in_sent > 0:
            oov_sentences.append(sentence)
            
    return oov_sentences

oov_sentences = sentence_with_oov(sentences, oov)
n_sentence_with_oov = len(oov_sentences)
```

```python
idx_shortest_oov_sentence, shortest_oov_sentence, len_shortest_oov_sentence = shortest_item(oov_sentences)
idx_longest_oov_sentence, longest_oov_sentence, len_longest_oov_sentence = longest_item(oov_sentences)

idx_shortest_oov_token, shortest_oov_token, len_shortest_oov_token = shortest_item(oov)
idx_longest_oov_token, longest_oov_token, len_longest_oov_token = longest_item(oov)

print(f"Number of OOV sentence: {n_sentence_with_oov} ({round(n_sentence_with_oov / n_sentence * 100, 2)}%)\n")

print(f"Index shortest oov sentence: {idx_shortest_oov_sentence}")
print(f"Shortest oov sentence: {shortest_oov_sentence}")
print(f"Length shortest sentence: {len_shortest_oov_sentence}\n")

print(f"Index longest oov sentence: {idx_longest_oov_sentence}")
print(f"Longest oov sentence: {longest_oov_sentence}")
print(f"Length longest oov sentence: {len_longest_oov_sentence}\n")

print(f"Index shortest oov token: {idx_shortest_oov_token}")
print(f"Shortest oov token: {shortest_oov_token}")
print(f"Length shortest token: {len_shortest_oov_token}\n")

print(f"Index longest oov token: {idx_longest_oov_token}")
print(f"Longest oov token: {longest_oov_token}")
print(f"Length longest oov token: {len_longest_oov_token}")
```

## Number of OOV each sentence

```python
def build_sentence(tokenize_sentences):
    sentences = []
    
    for sentence in tokenize_sentences:
        sentences.append("<SEP>".join(sentence))
        
    return sentences

sents = build_sentence(sentences)

def count_num_oov_each_sent(sentences, embedding):
    oovs = []
    
    for sentence in sentences:
        num_oov_in_sent = 0
            
        for token in sentence.split("<SEP>"):
            if token not in embedding:
                num_oov_in_sent += 1
                
        if num_oov_in_sent > 0:
            oovs.append(num_oov_in_sent)
        else:
            oovs.append(0)
        
    return oovs

oovs = count_num_oov_each_sent(sents, idn_word_embeddings)
idx_sentence_with_most_oov = oovs.index(max(oovs))

print(f"Number of OOV in sentence: {max(oovs)}")
print(sentences[idx_sentence_with_most_oov])
```
